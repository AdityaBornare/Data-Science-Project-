```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("disease_output.csv")
df = pd.DataFrame(data)

X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])

plt.bar(X, Y, color='r', align='edge')
plt.xticks(X, rotation='vertical')
plt.show()
```

```python
# Word Cloud
from wordcloud import WordCloud, STOPWORDS
fp = open("dis_symp_dict.txt", "r")
disease_file = fp.read()
stopwords = set(STOPWORDS)

disease_cloud = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

disease_cloud.generate(disease_file)

fig = plt.figure()
fig.set_figwidth(12) # set width
fig.set_figheight(15) # set height

plt.imshow(disease_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()

fp.close()
```
